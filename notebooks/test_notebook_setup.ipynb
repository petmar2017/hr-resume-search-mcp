```python
# Test 2: Import all required libraries
print("\nLibrary Import Test")
print("="*50)

import subprocess

def install_and_import(package, import_name=None):
    """Install package if needed and import it"""
    import_name = import_name or package
    try:
        module = __import__(import_name)
        print(f"‚úÖ {package}: {getattr(module, '__version__', 'OK')}")
        return module
    except ImportError:
        print(f"üì¶ Installing {package}...")
        subprocess.check_call([sys.executable, "-m", "pip", "install", package, "-q"])
        module = __import__(import_name)
        print(f"‚úÖ {package}: {getattr(module, '__version__', 'Installed')}")
        return module

# Test core dependencies
packages = [
    ("numpy", "numpy"),
    ("pandas", "pandas"),
    ("matplotlib", "matplotlib"),
    ("seaborn", "seaborn"),
    ("plotly", "plotly"),
    ("httpx", "httpx"),
    ("ipywidgets", "ipywidgets"),
    ("python-dotenv", "dotenv"),
]

for package, import_name in packages:
    install_and_import(package, import_name)
```

```python
# Test 5: Test async functionality
print("\nAsync Functionality Test")
print("="*50)

import asyncio

async def test_async():
    """Test async/await functionality"""
    await asyncio.sleep(0.1)
    return "Async works!"

try:
    # For Jupyter notebooks
    result = await test_async()
    print(f"‚úÖ Async/Await: {result}")
except:
    # Fallback for non-async context
    try:
        result = asyncio.run(test_async())
        print(f"‚úÖ Async/Await (via asyncio.run): {result}")
    except Exception as e:
        print(f"‚ùå Async/Await: {e}")
```

```python
# Test 6: Test API connectivity
print("\nAPI Connectivity Test")
print("="*50)

import httpx

async def test_api():
    """Test API connectivity"""
    async with httpx.AsyncClient() as client:
        try:
            # Test localhost API
            response = await client.get("http://localhost:8000/health", timeout=2.0)
            if response.status_code == 200:
                print("‚úÖ API Connection: FastAPI backend is running")
                print(f"   Response: {response.json()}")
            else:
                print(f"‚ö†Ô∏è API returned status {response.status_code}")
        except httpx.ConnectError:
            print("‚ÑπÔ∏è API not running (expected if not started)")
            print("   Start with: cd .. && make dev")
        except Exception as e:
            print(f"‚ùå API Connection Error: {e}")

try:
    await test_api()
except:
    asyncio.run(test_api())
```
